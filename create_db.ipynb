```python
import pandas as pd
from pathlib import Path
import os

db_path = "bike_store.db"

conn = sqlite3.connect(db_path)

folder = Path("data")
files = os.listdir(r"./data")
for file in files:
    table_name = file.split(".")[0]

    df = pd.read_csv(os.path.join("./data", file))

    df.to_sql(
        table_name,
        conn,
        if_exists="replace",
        index=False
    )

    print(f"Loaded {table_name}")

conn.close()
```
